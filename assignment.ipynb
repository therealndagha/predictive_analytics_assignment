<h2>Exploratory Data Analysis</h2>

The goal of EDA here is to describe your dataset using charts and statistics.
We want to answer questions like:

How many categories are there? Which are most common?

How many merchants? Who sells the most products?

How many products per cluster? Are there clusters with lots of offers?

What does Product Title text look like? (length, frequent words)

<h4>Category Distribution</h4>

```python
category_counts = df['Category Label'].value_counts()
print(category_counts)
```

```python
type(category_counts)
```

```python
plt.figure(figsize=(10,6))
plt.style.use("ggplot")
sns.barplot(x=category_counts.index, y=category_counts.values, hue=category_counts.index, legend=False)
plt.title("Number of Products Per Category")
plt.xticks(rotation=45)
plt.ylabel("Number of Products")
plt.xlabel("Category")
plt.show()
```
